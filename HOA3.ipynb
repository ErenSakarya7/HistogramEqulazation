```python
import cv2 as cv
from matplotlib import pyplot as plt
image=cv.imread('snow.png')
plt.imshow(image)
plt.show()

image_hsv=cv.cvtColor(image,cv.COLOR_BGR2HSV)

VChannel=image_hsv[:,:,2]
hist=cv.calcHist([VChannel],[0],None,[256],[0,256])
plt.imshow(VChannel,cmap='gray')
plt.title('V channel')
plt.show()
plt.hist(hist)
plt.title('Histogram of V channel')
plt.show()

equalized=cv.equalizeHist(VChannel)
plt.imshow(equalized,cmap='gray')
plt.title('Equalized V channel')
plt.show()
plt.hist(equalized)
plt.title('Histogram of equalized V channel')
plt.show()

HChannel=image_hsv[:,:,0]
SChannel=image_hsv[:,:,1]
VChannelEqualized=cv.equalizeHist(VChannel)
plt.imshow(cv.cvtColor(cv.merge([HChannel,SChannel,VChannelEqualized]),cv.COLOR_HSV2BGR))
plt.title('Equalized image')
plt.show()
```
